```python
#Coi như giới hạn n là 10^6
sangnt = [True]*(10**6+1)
n = int(input("Nhap so nguyen duong N: "))
for i in range(2, 10**6+1):
    if i>=2:
        for j in range(i*2, n+1, i):
            sangnt[j]=False
print(f"{n} {"" if sangnt[n] else "khong "}la so nguyen to.")
```

```python
n = int(input("Nhap so nguyen duong n: "))
sum = 0
for i in range(1, n+1):
    sum += i*(i+1)
print(f"Gia tri cua S(n) la: {sum}")
```

```python
a = list(map(int, input("Nhap day so nguyen: ").split()))
mx,mn = a[0], a[1]
for i in a:
    if i>mx:
        mx=i
    if i<mn:
        mn=i
print(f"Gia tri lon nhat trong day so la: {mx}")
print(f"Gia tri nho nhat trong day so la: {mn}")
```

```python
n = int(input("Nhap so nguyen duong n: "))
fibo = [0, 1, 1]
for i in range(3, n):
    fibo[0] = fibo[1]
    fibo[1] = fibo[2]
    fibo[2] = fibo[0] + fibo[1]
print(f"So Fibonacci thu {n} la: {fibo[2]}")
```

```python
def swap(a, b):
    return b, a
a = list(map(int, input("Nhap day so, cach nhau boi khoang trang: ").split()))
for i in range(0, len(a), 2):
    a[i], a[len(a)-1-i] = swap(a[i], a[len(a)-1-i])
print(f"Day so dao nguoc la: {a}")
```

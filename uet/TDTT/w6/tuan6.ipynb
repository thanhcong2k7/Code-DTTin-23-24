```python
#c3
def max_(a):
    tmp = a[0]
    for i in a:
        if i > tmp:
            tmp = i
    return tmp

n = list(map(int, input("Nhap mot day so: ").split()))
print(f"So lon nhat trong day: {max_(n)}")
```

```python
#c4
def min_(a):
    tmp = a[0]
    for i in a:
        if i < tmp:
            tmp = i
    return tmp

n = list(map(int, input("Nhap mot day so: ").split()))
print(f"So nho nhat trong day: {min_(n)}")
```

```python
#c5
sangnt = [False]*10001
sangnt[2] = True
for i in range(10001):
    if sangnt[i]:
        for j in range(i*2, 10001, i):
            sangnt[j] = False
def is_prime(n):
    return sangnt[n]

n = int(input("Nhap mot so nguyen duong: "))
print(f"{n} {"" if is_prime(n) else "khong"} la so nguyen to")
```

```python
#c9
def factorial(n):
    if n<2:
        return 1
    else:
        return factorial(n-1)*n
n = int(input("Nhap so nguyen duong: "))
print(f"{n}! = {factorial(n)}")
```

```python
#bai 1
n = int(input("Nhap so tu nhien N: "))
sangnt = [True]*10000
sangnt[0] = sangnt[1] = False
for i in range(2, int(10000**0.5)+1):
    if sangnt[i]:
        for j in range(i*i, 10000, i):
            sangnt[j] = False
for i in range(2,10000-2):
    if i>=n:
        break
    if sangnt[i] and sangnt[i+2]:
        print(f"{i} {i+2}")
```

```python
#bai 2
n = input("Nhap so tu nhien N: ")
sum = 0
for i in n:
    if int(i)%2 != 0:
        sum += int(i)
print(f"Tong cac chu so le la: {sum}")
```

```python
#bai 3
st = input("Nhap chuoi ky tu: ")
chr = input("Nhap 1 ky tu: ")
cnt = 0
for i in st:
    if i == chr:
        cnt += 1
print(f"Ky tu {chr} xuat hien {cnt} lan trong chuoi.")
```

```python
#bai 4 prefix sum
arr = list(map(int, input("Nhap day so nguyen: ").split()))
n = int(input("Nhap so tu nhien N: "))
mx = 0
for i in range(n):
    mx += arr[i]
for i in range(len(arr)-n+1):
    tmp = 0
    for j in range(i, i+n):
        tmp += arr[j]
    mx = max(mx, tmp)
    print(tmp)
print(f"Tong {n} so lien tiep lon nhat la: {mx}")
```

```python
#bai 5
arr = list(map(int, input("Nhap day so nguyen: ").split()))
chk = False
for i in range(int(len(arr)/2)):
    if arr[i] != arr[len(arr)-1-i]:
        print("Day so khong giong voi day dao nguoc")
        chk = True
        break
if not chk:
    print("Day so giong voi day dao nguoc")
```

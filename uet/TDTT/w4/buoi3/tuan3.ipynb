```python
def chk(a,b,c):
    if a+b>c:
        return True
    return False
a,b,c = map(float, input("Nhap 3 so nguyen: ").split())
if a>0 and b>0 and c>0:
    if chk(a,b,c) and (b,c,a) and chk(a,c,b) and chk(b,a,c) and chk(c,a,b) and chk(c,b,a):
        ncv = (a+b+c)/2
        print(f"Ba canh nay tao thanh mot tam giac.\nChu vi tam giac la: {ncv*2:.1f}\nDien tich tam giac la: {(ncv*(ncv-a)*(ncv-b)*(ncv-c))**0.5:.1f}")
```

```python
a,b,c = map(float, input("Nhap a, b, c: ").split())
d = b**2 - 4*a*c
if a==0 and b==0 and c==0:
    print("Phuong trinh vo so nghiem.")
elif a==0 and b==0:
    print("Phuong trinh vo nghiem.")
elif d==0:
    print(f"Phuong trinh co nghiem kep: {(-b-d**0.5)/(2*a):.1f}")
elif d<0:
    print("Phuong trinh vo nghiem.")
else:
    print(f"Phuong trinh co 2 nghiem rieng biet: {(-b-d**0.5)/(2*a):.1f} {(-b+d**0.5)/(2*a):.1f}")
```

```python
usage = int(input("Nhap so dien tieu thu (kWh): "))
bill = 0.0
if usage <= 50:
    bill=2000.0*(1 if usage>=50 else usage/50)
if 50 < usage <= 150:
    bill=2000.0 + 2500.0*(1 if usage>=150 else (usage-50)/100)
if 150 < usage <= 250:
    bill=2000.0 + 2500.0 + 3000.0*(1 if usage>=250 else (usage-150)/100)
if 250 < usage <= 350:
    bill=2000.0 + 2500.0 + 3000.0 + 3500.0*(1 if usage>=350 else (usage-250)/100)
if usage >= 350:
    bill=2000.0 + 2500.0 + 3000.0 + 3500.0 + 4000.0*((usage-350)/100)
print(f"Tien dien phai dong {bill:.1f}")
```
